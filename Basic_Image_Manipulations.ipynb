```
import cv2
#Uploaded a file for runtime
image = cv2.imread('image.jpg')
```

```
# cv2.imshow('original', image) # doesn't work in colab as it causes jupyter sessions to crash
# Hence using: from google.colab.patches import cv2_imshow
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
image.shape
```

```
# Resizing our image to 100 pixels wide 
r = 100/image.shape[1] # to remember the aspect ratio
dim = (100, int(image.shape[0]*r))

resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA) # The third argument is the algorithm to be used for resizing
cv2_imshow(resized)
```

```
# Lets' flip the image upside down i.e. rotate the image by 180 degrees around the center
# Grabbing the center of the image
(h,w) = image.shape[:2]
center = w/2, h/2
```

```
M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
cv2_imshow(rotated)
```

```
# Cropping the image
cropped = image[70:170, 440:540]
```

```
cv2_imshow(cropped)
```
